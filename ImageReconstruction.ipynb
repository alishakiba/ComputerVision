```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from skimage.transform import radon, rescale
```

```python
from skimage.data import shepp_logan_phantom
```

```python
image = shepp_logan_phantom()
plt.figure()
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
plt.close()
```

```python
theta1 = np.arange(0,171,10)
R1 = radon(image, theta1)

num_angles_R1 = R1.shape[1]
num_angles_R1
```

```python
theta2 = np.arange(0,176,5)
R2 = radon(image, theta2)

num_angles_R2 = R2.shape[1]
num_angles_R2
```

```python
theta3 = np.arange(0,179,2)
R3 = radon(image, theta3)

num_angles_R3 = R3.shape[1]
num_angles_R3
```

```python
N_R1 = R1.shape[0]
N_R2 = R2.shape[0]
N_R3 = R3.shape[0]
N_R1, N_R2, N_R3
```

```python
plt.figure(figsize=(8,5))
plt.imshow(R1, cmap=plt.cm.Greys_r, extent=(0, 180, 0, R1.shape[0]), aspect='auto')
plt.title('Radon transform\n({})'.format(num_angles_R1))
plt.xlabel(r'Parallel Rotation Angle - $\theta$ (degrees)')
plt.ylabel(r'Parallel Sensor Position - x (pixels)')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,5))
plt.imshow(R2, cmap=plt.cm.Greys_r, extent=(0, 180, 0, R2.shape[0]), aspect='auto')
plt.title('Radon transform\n({})'.format(num_angles_R2))
plt.xlabel(r'Parallel Rotation Angle - $\theta$ (degrees)')
plt.ylabel(r'Parallel Sensor Position - x (pixels)')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,5))
plt.imshow(R3, cmap=plt.cm.Greys_r, extent=(0, 180, 0, R3.shape[0]), aspect='auto')
plt.title('Radon transform\n({})'.format(num_angles_R3))
plt.xlabel(r'Parallel Rotation Angle - $\theta$ (degrees)')
plt.ylabel(r'Parallel Sensor Position - x (pixels)')
plt.show()
plt.close()
```

```python
from skimage.transform import iradon
```

```python
reconst_img_r1 = iradon(R1, theta1)
error_r1 = reconst_img_r1 - image
print(f"Filtered Back Projection rms reconstruction error: {np.sqrt(np.mean(error_r1**2)):.3g}")
```

```python
plt.figure()
plt.imshow(reconst_img_r1, cmap=plt.cm.gray)
plt.title('Reconstruction\nFiltered back projection')
plt.show()
plt.close()
```

```python
plt.figure()
plt.imshow(error_r1, cmap=plt.cm.gray)
plt.title('Reconstruction Error\nFiltered back projection')
plt.show()
plt.close()
```
